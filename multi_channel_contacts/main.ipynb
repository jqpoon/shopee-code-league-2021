```python
df.loc[df['Contacts'] ==0, 'No_Contacts'] = 'True'
# Replace each empty string with nan
df['Email'].replace('', np.nan, inplace=True)
df['Phone'].replace('', np.nan, inplace=True)
df['OrderId'].replace('', np.nan, inplace=True)
for i in range(len(df)):
    # for each index value, count the number of empty strings
    # if number of empty strings is 2 or more, then we remove that row
    count = df.loc[[i]].isna().sum().sum()
    if count >= 2 and df.iloc[i]['No_Contacts'] == 'True':
        print(i,count)
        df = df.drop(i)
print(len(df))
```

```python
df = pd.read_json("contacts.json")
# df.loc[df['Contacts'] ==0, 'No_Contacts'] = 'True'
# Replace each empty string with nan
df['Email'].replace('', np.nan, inplace=True)
df['Phone'].replace('', np.nan, inplace=True)
df['OrderId'].replace('', np.nan, inplace=True)
# Create new column with number of empty values
df['Missing'] = df.apply(lambda x: x.isna().sum(), axis='columns')
df.loc[df['Contacts'] ==0, 'No_Contacts'] = 'True'
df_new = df.drop(df[(df['Contacts'] == 0) & (df['Missing'] >= 2)].index)
```

```python
df_final = df_new.drop(columns = ['Missing', 'No_Contacts'])
print(df_final)
```

```python
df_final.to_csv('new_contacts.csv')
```
